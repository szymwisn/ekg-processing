## Zadanie 1
Wczytanie sygnałów z plików oraz wyświetlenie ich w formie interaktywnych wykresów.

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

# funkcja rysujaca wykres
def plot(time, amplitudes, file):
    fig = plt.figure(figsize=(8,5))
    ax = plt.axes()
    ax.set(xlabel='time [s]', ylabel='amplitude', title=file)
    ax.grid()
    ax.plot(time, amplitudes)
    plt.show()
```

`%matplotlib notebook` przywraca do Jupyter Notebooka interaktywność wykresów biblioteki matplotlib. Dzięki temu fragmentowi można je skalować, przesuwać, zapisywać wybrane fragmenty. Funkcjonalność przetestowana jedynie w Jupyter Notebook'u, brak gwarancji działania w innych środowiskach.

Funkcja rysująca wykres `plot(time, amplitudes, file)` przyjmuje 3 parametry - wektor czasu wyświetlony na osi X, wektor amplitud wyświetlony na osi Y oraz nazwę pliku, która wyświetli się jako tytuł wykresu.

### ekg_noise
1 kolumna: czas, 2 kolumna: wartości amplitud EKG, fs = 360 Hz

```python
file = './data/ekg_noise.txt'
data = np.loadtxt(file)
time = data[:,0]
amplitudes = data[:,1]

plot(time, amplitudes, file)
```

### ekg_1
12 kolumn odpowiada odprowadzeniom, fs = 1000 Hz

W pliku przechowywanych jest 12 różnych sygnałów (każda kolumna to osobny ciąg). Wybór kolumny następuje poprzez określenie zmiennej `column` (zakres 0-11).

Brakuje wektora określającego czas, w którym zostały dokonane poszczególne pomiary, więc został wygenerowany na podstawie częstotliwości próbkowania.

```python
column = 0

file = './data/ekg1.txt'
data = np.loadtxt(file)
measurements = data.shape[0]
amplitudes = data[:,column]
time = [x / 1000 for x in range(measurements)]

plot(time, amplitudes, file)
```

### ekg100.txt
1 kolumna, fs = 360 Hz

Podobnie jak w poprzednim pliku - również zabrakło wektora określającego czas, więc został wygenerowny na podstawie częśtotliwości próbkowania.

```python
file = './data/ekg100.txt'
data = np.loadtxt(file)
measurements = len(data)
amplitudes = data
time = [x / 360 for x in range(measurements)]

plot(time, amplitudes, file)
```
