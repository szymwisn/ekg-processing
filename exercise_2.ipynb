```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(time, amplitudes, title, xlabel, ylabel):
    fig = plt.figure(figsize=(25,10))
    ax = plt.axes()
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
    ax.grid()
    ax.plot(time, amplitudes)
    plt.show()
```

### 1. Wygeneruj ciąg próbek odpowiadający fali sinusoidalnej o częstotliwości 50 Hz i długości 65536.

```python
f = 50
length = 65536
fs = 1000

time = np.arange(length) / fs
amplitudes = np.sin(f * 2 * np.pi * time)

plot(time, amplitudes, f'Fala sinusoidalna {f} Hz o dlugosci {length}', 'time [s]', 'amplitude')
```

### 2. Wyznacz dyskretną transformatę Fouriera tego sygnału i przedstaw jego widmo amplitudowe na wykresie w zakresie częstotliwości [0, fs/2], gdzie fs oznacza częstotliwość próbkowania.

```python
fft = np.fft.fft(amplitudes) / length
fft_abs = np.abs(fft)
frequencies = fs * np.arange(length) / length

half = int(length/2)

plot(frequencies[0:half], fft_abs[0:half], f'Widmo sygnalu', 'frequency [Hz]', 'amplitude')
```

### 3. Wygeneruj ciąg próbek mieszaniny dwóch fal sinusoidalnych (tzn. ich kombinacji liniowej) o częstotliwościach 50 i 60 Hz. Wykonaj zadanie z punktu 2 dla tego sygnału.
